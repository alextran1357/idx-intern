```python
import subprocess
from pathlib import Path
```

```python
notebooks = [
    Path("week1_aggregation/aggregation.ipynb"),
    Path("week2-3_eda/listing_eda.ipynb"),
    Path("week2-3_eda/sold_eda.ipynb"),
    Path("week2-3_eda/mortgage.ipynb"),
    Path("week4-5_clean/listing_cleaup.ipynb"),
    Path("week4-5_clean/listing_cleaup.ipynb"),
    Path("week4-5_clean/sold_cleaup.ipynb"),
    Path("week6_feature_engineer/listing_feature_engineer.ipynb"),
    Path("week7_outlier/listing-outlier.ipynb"),
    Path("week7_outlier/sold-outlier.ipynb")
]
```

```python
for notebook in notebooks:
    subprocess.run(
        [
            "jupyter",
            "nbconvert",
            "--to",
            "notebook",
            "--execute",
            notebook.name,
            "--inplace"
        ],
        cwd=notebook.parent,
        check=True,
    )
```
