```python
import pandas as pd
from pathlib import Path
```

```python
LOAD_PATH = Path("../../data/combined")
```

```python
sold = pd.read_parquet(LOAD_PATH/"sold_fe.parquet")
```

```python
def print_IQR(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    

    print(column,"---------")
    print("Q1: ", Q1)
    print("Q3: ", Q3)
    print("IQR: ", IQR)
    print("lower: ", lower)
    print("upper: ", upper)
    print(df[column].describe(percentiles=[.01,.05,.1,0.9,0.95,0.99,0.995,0.999]))
    print()
```

```python
def percentile_columns(df, column, percentile):
    return df[column].quantile(percentile)
```

```python
# More obvious outliers
outliers = [
    'GarageSpaces',
    'MainLevelBedrooms',
    'BedroomsTotal',
    'BathroomsTotalInteger',
    'ParkingTotal'
]
for out in outliers:
    print_IQR(sold, out)
```

Looking at the IQR and percentages,

GarageSpaces, MainLevelBedrooms, BedroomsTotal, BathroomsTotalInteger, ParkingTotal numbers make a lot of sense at the 99th percentile. For these columns, everything after the 99th percentile will be removed

```python
sold.describe().T
```

```python
outliers = [
    'LotSizeArea',
    'AssociationFee',
    'LivingArea'
]
for out in outliers:
    print_IQR(sold, out)
```

```python
outliers = [
    'ClosePrice'
]
for out in outliers:
    print_IQR(sold, out)
```

```python
outlier_sold = sold.copy()
outlier_sold["Flagged"] = False

condition = outlier_sold["GarageSpaces"].gt(percentile_columns(outlier_sold,"GarageSpaces",.999)).fillna(False)
outlier_sold["Flagged"] = outlier_sold["Flagged"] | condition
condition = outlier_sold["MainLevelBedrooms"].gt(percentile_columns(outlier_sold,"MainLevelBedrooms",.999)).fillna(False)
outlier_sold["Flagged"] = outlier_sold["Flagged"] | condition
condition = outlier_sold["BedroomsTotal"].gt(percentile_columns(outlier_sold,"BedroomsTotal",.999)).fillna(False)
outlier_sold["Flagged"] = outlier_sold["Flagged"] | condition
condition = outlier_sold["BathroomsTotalInteger"].gt(percentile_columns(outlier_sold,"BathroomsTotalInteger",.999)).fillna(False)
outlier_sold["Flagged"] = outlier_sold["Flagged"] | condition
condition = outlier_sold["ParkingTotal"].gt(percentile_columns(outlier_sold,"ParkingTotal",.999)).fillna(False)
outlier_sold["Flagged"] = outlier_sold["Flagged"] | condition

condition = outlier_sold["LotSizeArea"].gt(percentile_columns(outlier_sold,"LotSizeArea",.999)).fillna(False)
outlier_sold["Flagged"] = outlier_sold["Flagged"] | condition
condition = outlier_sold["AssociationFee"].gt(percentile_columns(outlier_sold,"AssociationFee",.999)).fillna(False)
outlier_sold["Flagged"] = outlier_sold["Flagged"] | condition
condition = outlier_sold["LivingArea"].gt(percentile_columns(outlier_sold,"LivingArea",.999)).fillna(False)
outlier_sold["Flagged"] = outlier_sold["Flagged"] | condition

condition = outlier_sold["ClosePrice"].le(0).fillna(False)
outlier_sold["Flagged"] = outlier_sold["Flagged"] | condition
condition = outlier_sold["ClosePrice"].gt(2415500.0).fillna(False)
outlier_sold["Flagged"] = outlier_sold["Flagged"] | condition

print(outlier_sold[outlier_sold["Flagged"]].shape[0]/ outlier_sold.shape[0])
```

```python
sold_removed = outlier_sold.copy()

sold_removed = sold_removed.drop(sold_removed[sold_removed['Flagged']==True].index)
```
