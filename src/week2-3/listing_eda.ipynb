## ClosePrice

```python
fig, ax = plt.subplots(2, 1, figsize=(8,8))
ax[0].hist(listing['ClosePrice'], bins=50)
ax[0].set_yscale('log')
ax[0].set_title('ClosePrice Histogram (log scale)')

ax[1] = listing['ClosePrice'].plot(kind="box", logx=True, vert=False)
ax[1].set_title('ClosePrice Boxplot')

plt.tight_layout()
plt.show()
listing['ClosePrice'].describe(percentiles=[0.25,0.5,0.75,0.9,0.95,0.99])
```

The boxplot shows a high number of outliers. Looking at the percentiles, 99% of the data is priced below 5.5m while the max is 820m.
