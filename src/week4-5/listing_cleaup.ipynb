```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
save_path = Path("../../data/combined")
```

```python
listing = pd.read_csv(save_path/"listing_fred.csv")
```

```python
market_analysis_fields = ['OriginalListPrice'
,'CloseDate'
,'ClosePrice'
,'Latitude'
,'Longitude'
,'PropertyType'
,'LivingArea'
,'ListPrice'
,'DaysOnMarket'
,'FireplacesTotal'
,'AssociationFeeFrequency'
,'AboveGradeFinishedArea'
,'TaxAnnualAmount'
,'CountyOrParish'
,'ElementarySchool'
,'AttachedGarageYN'
,'ParkingTotal'
,'PropertySubType'
,'LotSizeAcres'
,'YearBuilt'
,'BathroomsTotalInteger'
,'City'
,'BuildingAreaTotal'
,'BedroomsTotal'
,'BelowGradeFinishedArea'
,'StateOrProvince'
,'CoveredSpaces'
,'FireplaceYN'
,'Stories'
,'HighSchool'
,'Levels'
,'LotSizeDimensions'
,'LotSizeArea'
,'MainLevelBedrooms'
,'NewConstructionYN'
,'GarageSpaces'
,'PostalCode'
,'AssociationFee'
,'LotSizeSquareFeet'
,'PurchaseContractDate'
,'ListingContractDate'
,'ContractStatusChangeDate'
,'MiddleOrJuniorSchoolDistrict']
```

```python
# Filter by Market Analysis Fields
market_analysis_filter_listing = listing[market_analysis_fields]
```

```python
market_analysis_filter_listing.columns
```

```python
null_90_columns = [
    'MiddleOrJuniorSchoolDistrict',
    'CoveredSpaces',
    'TaxAnnualAmount',
    'FireplacesTotal',
    'AboveGradeFinishedArea',
    'BelowGradeFinishedArea',
    'LotSizeDimensions',
]
```

```python
# Filter by dropping above 90% null rate
null_90_filter = market_analysis_filter_listing.drop(columns=null_90_columns)
```

```python
null_90_filter.columns
```

```python
null_90_filter.shape
```

```python
null_90_filter.dtypes
```

```python
null_90_filter['CloseDate']
```

```python
# Datetime Change
null_90_filter['CloseDate'] = pd.to_datetime(null_90_filter['CloseDate'])
null_90_filter['PurchaseContractDate'] = pd.to_datetime(null_90_filter['PurchaseContractDate'])
null_90_filter['ListingContractDate'] = pd.to_datetime(null_90_filter['ListingContractDate'])
null_90_filter['ContractStatusChangeDate'] = pd.to_datetime(null_90_filter['ContractStatusChangeDate'])

# Int change
null_90_filter['YearBuilt'] = pd.to_numeric(null_90_filter['YearBuilt'])
null_90_filter['BathroomsTotalInteger'] = pd.to_numeric(null_90_filter['BathroomsTotalInteger'])
null_90_filter['BedroomsTotal'] = pd.to_numeric(null_90_filter['BedroomsTotal'])
null_90_filter['Stories'] = pd.to_numeric(null_90_filter['Stories'])
null_90_filter['MainLevelBedrooms'] = pd.to_numeric(null_90_filter['MainLevelBedrooms'])

# Bool change
bool_cols = ['AttachedGarageYN','FireplaceYN','NewConstructionYN']
null_90_filter[bool_cols] = null_90_filter[bool_cols].astype(bool)
```
