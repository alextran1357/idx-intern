```python
import pandas as pd
import numpy as np
from pathlib import Path
import matplotlib.pyplot as plt
import datetime as dt
```

```python
def print_missing_data(df):
        
    # percent missing data per column
    missing_count = df.isnull().sum()
    percent_missing = round(missing_count * 100 / len(df), 2)
    missing_value_df = pd.DataFrame(
        {
            'missing_count': missing_count,
            'percent_missing': percent_missing
        }
    )
    missing_value_df.sort_values('percent_missing', inplace=True)
    return missing_value_df
```

```python
# Filter by Market Analysis Fields
market_analysis_filter_listing = listing[market_analysis_fields]
```

```python
market_analysis_filter_listing.columns
```

```python
# drop columns

null_90_columns = [
    'MiddleOrJuniorSchoolDistrict',
    'CoveredSpaces',
    'TaxAnnualAmount',
    'FireplacesTotal',
    'AboveGradeFinishedArea',
    'BelowGradeFinishedArea',
    'LotSizeDimensions',
    'BuildingAreaTotal',
    # I originally wanted to keep these to check for nearby schools. but there is too many missing data to check for t-tests
    'HighSchool',
    'ElementarySchool',
]
dup_columns = [
    'LotSizeAcres', 
    'LotSizeSquareFeet', # There is LotSizeArea which is the same as 'feet' and it has less missing data
    'Stories', # Stories(float) and Levels(category) are the same. Stories will be null if levels is more than 2 stories high
    'PropertyType', # We already filtered by 'Residential' 
]
```

```python
# Filter by dropping above 90% null rate
null_90_filter = market_analysis_filter_listing.drop(columns=null_90_columns)
null_90_filter = null_90_filter.drop(columns=dup_columns)
```

```python
null_90_filter.columns
```

```python
null_90_filter.dtypes
```

```python
# Datetime Change
null_90_filter['CloseDate'] = pd.to_datetime(null_90_filter['CloseDate'])
null_90_filter['PurchaseContractDate'] = pd.to_datetime(null_90_filter['PurchaseContractDate'])
null_90_filter['ListingContractDate'] = pd.to_datetime(null_90_filter['ListingContractDate'])
null_90_filter['ContractStatusChangeDate'] = pd.to_datetime(null_90_filter['ContractStatusChangeDate'])

# Int change
int_cols = ['YearBuilt','BathroomsTotalInteger','BedroomsTotal', 'MainLevelBedrooms']
null_90_filter[int_cols] = null_90_filter[int_cols].astype('Int64')

# Bool change
bool_cols = ['FireplaceYN','NewConstructionYN']
null_90_filter[bool_cols] = null_90_filter[bool_cols].astype(bool)
```

```python
null_90_filter.dtypes
```

```python
# Check date consistency
null_90_filter['listing_after_close_flag'] = (null_90_filter['CloseDate'] - null_90_filter['ListingContractDate']< dt.timedelta(days=0))
null_90_filter['purchase_after_close_flag'] = (null_90_filter['CloseDate'] - null_90_filter['PurchaseContractDate']< dt.timedelta(days=0))
null_90_filter['geographic_error'] = (
    (null_90_filter["Longitude"]==0) |
    (null_90_filter["Latitude"]==0) |
    (null_90_filter["Longitude"]>0) |
    (null_90_filter["Longitude"].isna()) |
    (null_90_filter["Latitude"].isna()) 
)
```

```python
print_missing_data(null_90_filter)
```

```python
# Drops rows

# negative
negative_filter = null_90_filter.drop(null_90_filter[null_90_filter['DaysOnMarket'] < 0].index)
negative_filter = negative_filter.drop(negative_filter[negative_filter['LivingArea'] <= 0].index)
negative_filter = negative_filter.drop(negative_filter[negative_filter['ClosePrice'] <= 0].index)
negative_filter = negative_filter.drop(negative_filter[negative_filter['BedroomsTotal'] <= 0].index)
negative_filter = negative_filter.drop(negative_filter[negative_filter['BathroomsTotalInteger'] <= 0].index)

# inconsistent
inconsistent_filter = negative_filter.drop(negative_filter[negative_filter['PostalCode']=="0"].index) # cannot be 0
inconsistent_filter = inconsistent_filter.drop(inconsistent_filter[inconsistent_filter['StateOrProvince']!="CA"].index) # outside of CA

# null 
# null_filter = inconsistent_filter.drop(inconsistent_filter[inconsistent_filter['PostalCode'].isna()].index) # postalcode nulls 
```
