```python
import pandas as pd
import glob
from pathlib import Path
import chardet
```

```python
data_path = Path("../../data/raw")
save_path = Path("../../data/combined")
```

```python
listing_file = list(data_path.glob("CRMLSListing*.csv"))
sold_file = list(data_path.glob("CRMLSSold*.csv"))
```

```python
full_listing = []
encoding_log = []
row_count_listing = []

for filename in listing_file:
    try:
        df = pd.read_csv(filename, encoding="utf-8", low_memory=False)
        encoding_log.append((filename, "utf-8"))
    except UnicodeDecodeError:
        df = pd.read_csv(filename, encoding="cp1252", low_memory=False)
        encoding_log.append((filename, "cp1252"))
        
    row_count_listing.append(len(df))
    full_listing.append(df)
listing_combined = pd.concat(full_listing, axis=0, ignore_index=True)
```

```python
encoding_log
```

```python
full_sold = []
encoding_log = []
row_count_sold = []
for filename in sold_file:
    try:
        df = pd.read_csv(filename, encoding="utf-8", low_memory=False)
        encoding_log.append((filename, "utf-8"))
    except UnicodeDecodeError:
        df = pd.read_csv(filename, encoding="cp1252", low_memory=False)
        encoding_log.append((filename, "cp1252"))
        
    row_count_sold.append(len(df))
    full_sold.append(df)
sold_combined = pd.concat(full_sold, axis=0, ignore_index=True)
```

```python
encoding_log
```

# Deliverable

## Row Counts

```python
len(listing_combined) == sum(row_count_listing)
```

```python
len(sold_combined) == sum(row_count_sold)
```

## Filter

```python
# Before Filter

print(sum(row_count_listing))
print(sum(row_count_sold))
```

```python
resident_listing_filter = listing_combined[listing_combined["PropertyType"]=="Residential"]
```

```python
resident_sold_filter = sold_combined[sold_combined["PropertyType"]=="Residential"]
```

```python
print(len(resident_listing_filter))
print(len(resident_sold_filter))
```

```python
resident_listing_filter.to_csv(save_path/"listing.csv", encoding="utf-8", index=False)
resident_sold_filter.to_csv(save_path/"sold.csv", encoding="utf-8", index=False)
```
